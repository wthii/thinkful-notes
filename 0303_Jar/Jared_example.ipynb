## cars data set

```python
cars = pd.read_csv('mtcars.csv')
```

```python
cars.columns
```

```python
cars.groupby('cyl')['mpg'].mean()
```

```python
cars.groupby('cyl')['mpg'].mean().plot()
```

```python
cars.groupby('cyl')['mpg'].mean().plot(style='.')
```

```python
def split_model(model):
    return model.split()[0]

cars['manufactoror'] = cars['model'].apply(split_model)
```

```python
cars.groupby('manufactoror')['mpg'].mean().sort_values()
```
